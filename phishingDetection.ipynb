```python
import pandas as pd
```

```python
ds=pd.read_csv('dataset_website.csv')
ds
```

```python
ds.info()
ds.describe()
```

```python
ds=ds.drop('index',axis=1)
ds
```

```python
ds.count()
```

```python
y=ds['Result']
x=ds.drop('Result',axis=1)
x,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
x_test,x_train,y_train,y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=4)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=model.predict(x_test)
accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.tree import DecisionTreeClassifier
dmodel=DecisionTreeClassifier()
dmodel.fit(x_train,y_train)
y_pred=dmodel.predict(x_test)
accuracy_score(y_test,y_pred)*100
```

```python
from sklearn.tree import export_graphviz
export_graphviz(dmodel,out_file='classificationTree.dot')
```

```python
from sklearn.ensemble import RandomForestClassifier
rmodel=RandomForestClassifier()
rmodel.fit(x_train,y_train)
y_pred=rmodel.predict(x_test)
accuracy_score(y_pred,y_test)*100
```

```python
from sklearn.svm import SVC
smodel=SVC(kernel='poly')
smodel.fit(x_train,y_train)
y_pred=smodel.predict(x_test)
accuracy_score(y_pred,y_test)*100
```

```python
from sklearn import metrics
from matplotlib import pyplot as plt
cmatrix=metrics.confusion_matrix(y_pred,y_test)
cmatrix_display=metrics.ConfusionMatrixDisplay(confusion_matrix=cmatrix)
cmatrix_display.plot()
cmatrix,plt.show()
```
